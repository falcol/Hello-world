## V·∫Ω ƒë·ªì th·ªã h√†m x = y**2 v√† x = y tr√™n c√πng 1 tr·ª•c to·∫° ƒë·ªô.

```python
y = np.linspace(0, 10, num=100)
plt.style.use('ggplot')
plt.plot(y**2 , y)
plt.plot(y, y)
plt.title('ƒê·ªì th·ªã h√†m x = y**2 v√† x = y')
plt.legend(['x = y**2', 'x = y'])
```
